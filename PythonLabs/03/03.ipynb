```python
# Медведев Георгий; лаб.3, упр.3

import turtle as t

size = 30                             
diag_size = size * (2**0.5)           
gap = 10                              

def truth(flag):                      
    if int(flag) == 1:                #теперь считывается строка, поэтому делаем привидение типа
        t.pendown()
    else:
        t.penup()

def graf(pupds):                      #функция обхода общего контура цифр
    truth(pupds[0])
    t.right(180)
    t.forward(size)
        
    truth(pupds[1])
    t.left(90)
    t.forward(size)
        
    truth(pupds[2])
    t.left(135)
    t.forward(diag_size)
        
    truth(pupds[3])
    t.right(135)
    t.forward(size)
        
    truth(pupds[4])
    t.right(90)
    t.forward(size)
        
    truth(pupds[5])
    t.left(90)
    t.forward(size)
        
    truth(pupds[6])
    t.left(135)
    t.forward(diag_size)
        
    truth(pupds[7])
    t.right(135)
    t.forward(size)
        
    truth(pupds[8])
    t.right(90)
    t.forward(size)
    
    t.penup()
    

inp = open('input.txt', 'r')        #открытие файла (прикреплен к задаче)
A = inp.readlines()                 

for i in range(6):
    t.goto(i*(size + gap), 0)
    t.setheading(0)
    graf(A[i])

inp.close()

t.exitonclick()
```
