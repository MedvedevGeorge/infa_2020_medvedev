```python
# Медведев Георгий; лаб.6 + 7

import pygame
from pygame.draw import *
from random import randint
import numpy as np
pygame.init()

FPS = 10                                            #частота кадров
screen = pygame.display.set_mode((800, 600))        #размер окна

RED = (255, 0, 0)                                   #список цветов
BLUE = (0, 0, 255)
YELLOW = (255, 255, 0)
GREEN = (0, 255, 0)
MAGENTA = (255, 0, 255)
CYAN = (0, 255, 255)
COLORS = [RED, BLUE, YELLOW, GREEN, MAGENTA, CYAN]
TITLE = (255, 140, 0)
BLACK = (0, 0, 0)
WHITE = (255, 255, 255)

def title_screen():
    '''
    Отрисовывает экран заставки с названием игры: "MEGA AIM"
    '''
    #M
    polygon(screen, TITLE, [(20, 260), (110, 50), (140, 130), (195, 45), 
                            (260, 265), (200, 240), (170, 160), (150, 215),
                            (125, 170), (80, 250), (20, 260)])

    #E
    polygon(screen, TITLE, [(220, 50), (345, 40), (320, 80), (265, 85), 
                            (295, 135), (370, 115), (385, 180), (310, 175),
                            (315, 240), (415, 215), (440, 265), (290, 270), (220, 50)]) 

    #G
    polygon(screen, TITLE, [(380, 50), (680, 45), (660, 90), (480, 100), 
                            (510, 225), (590, 230), (585, 200), (555, 205),
                            (525, 170), (590, 160), (640, 260), (480, 265), (380, 50)]) 

    #A + вырез
    polygon(screen, TITLE, [(650, 270), (710, 50), (790, 265), (755, 275), 
                            (745, 230), (700, 235), (700, 270), (650, 270)]) 
    polygon(screen, BLACK, [(700, 205), (705, 125), (740, 205), (700, 205)])

    #A + вырез
    polygon(screen, TITLE, [(165, 440), (250, 290), (335, 445), (290, 440), 
                            (280, 410), (210, 415), (200, 440), (165, 440)]) 
    polygon(screen, BLACK, [(225, 385), (250, 335), (270, 385), (225, 385)])

    #I
    polygon(screen, TITLE, [(300, 295), (515, 295), (500, 325), (430, 320), 
                            (460, 420), (545, 415), (555, 445), (365, 450), 
                            (345, 420), (420, 420), (390, 320), (310, 325), (300, 295)]) 

    #M
    polygon(screen, TITLE, [(565, 440), (530, 295), (630, 360), (710, 290), 
                            (700, 480), (675, 440), (670, 385), (630, 420), 
                            (570, 370), (595, 440), (565, 440)]) 

    #автор
    font = pygame.font.Font(None, 50)                                   
    text = font.render("a game by George Medvedev", True, WHITE)
    screen.blit(text, (185, 515))    

    #реквизиты
    font = pygame.font.Font(None, 50)                                   
    text = font.render("MIPT 2020", True, WHITE)
    screen.blit(text, (330, 555))  

    pygame.display.update()
    
def new_ball():
    '''
    Создаёт новый шарик с параметрами: 
        1) цвет: случайный цвет из списка цветов
        2) радиус: случайный целый радиус из диапазона [10, 30]
        3) координаты центра: случайные целые числа в пределах 
                экрана с отступом в 30 от границ, чтобы шар с максимальным 
                радиусом заведомо влез на экран полностью
        4) скорость центра шара: проекции по осям X и Y - случайные 
                числа не превосходящие по модулю 10
    А также рисует этот шарик
    Шарик будет двигаться прямолинейно с упругим отражением от стен
    '''
    global x, y, r, color, v_x, v_y
    x = randint(30, 770)
    y = randint(30, 570)
    r = randint(10, 30)
    color = COLORS[randint(0, 5)]
    v_x = randint(-10, 10)
    v_y = randint(-10, 10)
    circle(screen, color, (x, y), r)
    
def new_aim():
    '''
    Создаёт новую мишень с параметрами: 
        1) цвет: случайный цвет из списка цветов
            1.1) маленький цвет: цвет внутренней мешени
        2) радиус: случайный целый радиус из диапазона [10, 30]
            2.1) маленький радиус: радиус внутренней мешени
        3) координаты центра: случайные целые числа в пределах 
                экрана с отступом в 30 от границ, чтобы шар с максимальным 
                радиусом заведомо влез на экран полностью
    А также рисует эту мишень
    Мишень будет двигаться по дуге окружности со случайным направлением
    При клике по внутренней мешени будут выдаваться бонусные очки
    '''
    global x_aim, y_aim, r_aim, R_aim, color_aim_1, color_aim_2
    x_aim = randint(100, 700)
    y_aim = randint(100, 500)
    R_aim = randint(10, 30)
    r_aim = R_aim // 3
    n = randint(0, 5)
    color_aim_1 = COLORS[n]
    color_aim_2 = COLORS[5 - n]
    circle(screen, color_aim_1, (x_aim, y_aim), R_aim)
    circle(screen, color_aim_2, (x_aim, y_aim), r_aim)

def click(event):
    '''
    Обрабатывает клики по шарику и по бонусной мишени
    Возвращает количество очков в зависимости от радиуса и типа цели
    Можно абузить проклики (!FIXME)
    '''
    distance = ((event.pos[0] - x_aim)**2 + (event.pos[1] - y_aim)**2)**0.5 
    if distance <= r_aim:
        return 5*(35 - R_aim)
    elif distance <= R_aim:
        return (35 - R_aim)
    for i in range(num):
        distance = ((event.pos[0] - X[i])**2 + (event.pos[1] - Y[i])**2)**0.5 
        if distance <= R[i]:
            return (35 - R[i])
    return 0
    
pygame.display.update()
clock = pygame.time.Clock()
finished = False
points = 0  
num = 3                                          #одновременное количество шаров на экране
X = [0]*num                                      #списки для хранения данных о шарах
Y = [0]*num
V_x = [0]*num
V_y = [0]*num
R = [0]*num
Col = [0]*num

player_name = input("Как вас зовут? ")           #запись имени игрока

title_screen()                                   #запуск заставки
pygame.time.wait(3000)                           #ждём чтобы игрок успел прочитать заставку

while not finished:
    screen.fill(BLACK)
    for i in range (num):                        #создаем "num" шаров и записываем их параметры
        new_ball()
        X[i] = x
        Y[i] = y
        V_x[i] = v_x
        V_y[i] = v_y
        R[i] = r
        Col[i] = color

    new_aim()                                    #создаем бонусную мишень 
    
    font = pygame.font.Font(None, 30)            #табло очков
    text = font.render('Score: ' + str(points), True, WHITE)
    screen.blit(text, (0, 0)) 
    
    pygame.display.update() 
    clock.tick(FPS)
    for event in pygame.event.get():             #диспетчеризация событий
        if event.type == pygame.QUIT:
            finished = True
        elif event.type == pygame.MOUSEBUTTONDOWN:
            points += click(event)
                
    #загатовка параметров для движения бонусной мишени: ось вращения + направление
    rot_x = x_aim + randint(30, 50)               
    rot_y = y_aim + randint(30, 50) 
    if randint(0, 1) == 0:
        a = 0.1 
    else:
        a = -0.1 

    for t in range(30):                          #движение шаров и бонусной мишени
        screen.fill(BLACK)                       #пока все шары на экране двигаются одинаковое время (!FIXME)
        for i in range(num):
            X[i] += V_x[i]
            Y[i] += V_y[i]
            if X[i] < 30 or X[i] > 770:          #обработка столкновений со стенами
                V_x[i] = -V_x[i]
            elif Y[i] < 30 or Y[i] > 570:
                V_y[i] = -V_y[i]
            circle(screen, Col[i], (X[i], Y[i]), R[i])

        #поворот бонусной мишени
        x_tmp = (x_aim - rot_x)*np.cos(a) - (y_aim - rot_y)*np.sin(a) + rot_x
        y_tmp = (x_aim - rot_x)*np.sin(a) + (y_aim - rot_y)*np.cos(a) + rot_y
        x_aim = int(x_tmp)
        y_aim = int(y_tmp)
        circle(screen, color_aim_1, (x_aim, y_aim), R_aim)
        circle(screen, color_aim_2, (x_aim, y_aim), r_aim)
        
        font = pygame.font.Font(None, 30)        #табло очков
        text = font.render('Score: ' + str(points), True, WHITE)
        screen.blit(text, (0, 0)) 

        pygame.display.update() 
        clock.tick(FPS)
        for event in pygame.event.get():         #диспетчеризация событий
            if event.type == pygame.QUIT:
                finished = True
            elif event.type == pygame.MOUSEBUTTONDOWN:
                points += click(event)


out = open('High score.txt', 'a')                #запись набранного количества очков (файл прикреплен к задаче)      
out.write(player_name + ' - ' + str(points) + '\n')     
out.close()
                    
pygame.quit()
```
