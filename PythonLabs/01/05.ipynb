```python
# Медведев Георгий; лаб.1, упр.5

import numpy as np
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5, 6]
y = [1, 1.42, 1.76, 2, 2.24, 2.5]
p1, v1 = np.polyfit(x, y, deg=1, cov=True)
p2, v2 = np.polyfit(x, y, deg=2, cov=True)

p_f1d = np.poly1d(p1)
p_f2d = np.poly1d(p2)

plt.plot(x, p_f1d(x), 'b', x, p_f2d(x), 'y') # график линейного приближения - синим, и квадратичного - желтым
plt.errorbar(x, y, xerr=0.1, yerr=0.1, fmt='ro') # точки данных с крестами погрешностей красным цветом (погрешнось по x сделал побольше для наглядности)
plt.grid()
plt.show()
```
